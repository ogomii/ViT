```python
import torch
```

```python
def attn(x):
    print(x.shape)
    assert len(x.shape) == 2 # B, T
    return x
```

```python
a = torch.tensor([[i for i in range(1,17)] for i in range(6)])
B, N = a.shape
patch_size = 4
window_size = 2
a = a.reshape(B, patch_size//window_size, window_size, patch_size//window_size, window_size).permute(0,1,3,2,4).reshape(B,-1,window_size,window_size)
for i in range(a.shape[1]):
    a[:, i, :, :] = a[:, i, :, :] + attn(a[:, i, :, :].flatten(start_dim=1,end_dim=2)).reshape(B,window_size, window_size)
a = a.reshape(B, 2,2,2,2).permute(0,1,3,2,4).reshape(B, 4,4).flatten(start_dim=1,end_dim=2)
a[1] += 1
a
```

```python
from torch.nn import functional as F
a = a.permute(1,0)
F.avg_pool1d(a, kernel_size=6).shape
```
